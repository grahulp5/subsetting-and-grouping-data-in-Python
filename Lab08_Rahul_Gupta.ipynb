### Task 2 (0.5 mark)

Can you use conditional selection and Boolean filtering to figure out the mean age of those who responded __'Yes'__ and __'No'__ to the following questions:

```Have you ever tried to meet up with hometown friends on Thanksgiving night?```

```Have you ever attended a "Friendsgiving?"```

You should be getting the following answers for the different combinations:

    'No' and 'Yes': 37.01052631578948

    'Yes' and 'No': 41.47540983606557

````python
df=thanksgiving_data
df.head()
````

````python
no_no_filter = (df.iloc[:,56] == "No") & (df.iloc[:,57] == "No")
no_yes_filter = (df.iloc[:,56] == "No") & (df.iloc[:,57] == "Yes")
yes_no_filter = (df.iloc[:,56] == "Yes") & (df.iloc[:,57] == "No")
yes_yes_filter = (df.iloc[:,56] == "Yes") & (df.iloc[:,57] == "Yes")

mean_age_no_no = df[no_no_filter]["int_age"].mean()
mean_age_no_yes = df[no_yes_filter]["int_age"].mean()
mean_age_yes_no = df[yes_no_filter]["int_age"].mean()
mean_age_yes_yes = df[yes_yes_filter]["int_age"].mean()

print(f"'No' and 'No': {mean_age_no_no:}")
print(f"'No' and 'Yes': {mean_age_no_yes:}")
print(f"'Yes' and 'No': {mean_age_yes_no:}")
print(f"'Yes' and 'Yes': {mean_age_yes_yes:}")
````

### Task 4 (1 mark)

Can you get the same results you got in __Task 3__ using the __groupby__ method? A snapshot of the expected result is shown below.

<img align="left" src="images/task4.png" height="800" width="800">

````python
### Write your code below this comment.
grouped = df.groupby([df.iloc[:, 56], df.iloc[:, 57]])['int_age'].mean()
grouped
````

````python
formatted_data = grouped.unstack().T
formatted_data = formatted_data.stack().reset_index()
formatted_data['int_age'] = formatted_data[0]
formatted_data.drop(0, axis=1, inplace=True)
formatted_data.columns = [df.columns[56], df.columns[57], 'Mean Age']

print(formatted_data.to_string(index=False))
````
